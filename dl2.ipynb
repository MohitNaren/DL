```python
import numpy as np
def step(x):
  return 1 if x >= 0 else 0
class Perceptron:
  def __init__(self, weights, bias):
    self.weights = weights
    self.bias = bias
  def predict(self, inputs):
    t = np.dot(self.weights, inputs) + self.bias
    return step(t)
#AND Gate parameters
weights = np.array([1,1])
bias = -1.5
a_g = Perceptron(weights, bias)
print("AND Gate")
for x in [(0,0), (0,1), (1,0), (1,1)]:
  print(x, "->", a_g.predict(np.array(x)))
```

```python
weights = np.array([1,1])
bias = -0.5
#OR Gate parameters
o_g = Perceptron(weights, bias)
print("OR Gate")
for x in [(0,0), (0,1), (1,0), (1,1)]:
  print(x, "->", o_g.predict(np.array(x)))
```

```python
def step(x):
  return 1 if x>= 0 else 0
def xor(x1, x2):
  p1 = step(x1 + x2 - 0.5)
  p2 = step(x1 + x2 - 1.5)
  return step(p1 - 2*p2 -  0.5)
for a,b in [(0,0), (0,1), (1,0), (1,1)]:
  print((a, b), "->", xor(a, b))
```

```python
import numpy as np
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x):
    return x * (1 - x)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y = np.array([[1], [0], [0], [1]])
np.random.seed(42)
W1 = np.random.rand(2, 2)
b1 = np.random.rand(1, 2)
W2 = np.random.rand(2, 1)
b2 = np.random.rand(1, 1)
learning_rate = 0.1
epochs = 10000
for _ in range(epochs):
    hidden = sigmoid(np.dot(X, W1) + b1)
    y_pred = sigmoid(np.dot(hidden, W2) + b2)
    error = y - y_pred
    d_out = error * sigmoid_derivative(y_pred)
    d_hidden = d_out.dot(W2.T) * sigmoid_derivative(hidden)
    W2 += hidden.T.dot(d_out) * learning_rate
    b2 += np.sum(d_out, axis=0, keepdims=True) * learning_rate
    W1 += X.T.dot(d_hidden) * learning_rate
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate
for i in range(len(X)):
    print(X[i], "->", round(y_pred[i][0]))
```
